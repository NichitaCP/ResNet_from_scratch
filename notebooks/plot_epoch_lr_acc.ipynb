```python
import pandas as pd 
import matplotlib.pyplot as plt 
import os
```

```python
df = pd.read_csv("../model_results/epoch_lrs.csv")
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2, 2, 1)
plt.plot(df.Epoch, df.Lr, label="Learning rate")
plt.legend()
plt.subplot(2, 2, 2)
plt.plot(df.Epoch, df.TrLoss, c='r', label="Train Loss")
plt.plot(df.Epoch, df.ValLoss, '--', c='g', label="Validation Loss")
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(df.Epoch, df.TrAcc, c='black', label="Train Accuracy")
plt.plot(df.Epoch, df.ValAcc, c='cyan', label="Validation Accuracy")
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(df.Epoch, df.TrROC, label="Train ROC-AUC")
plt.plot(df.Epoch, df.ValRoc, label="Validation ROC-AUC")
plt.legend()
plt.savefig("../plots/train_validation_results.png")
plt.show()
```
